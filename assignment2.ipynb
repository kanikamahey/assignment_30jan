the WHILE LOOP  is the condition controlled loop that lets you execute statements repeatedly as long as boolean expression
specifying test cndition remains true .

it is mostly used in case when the programmer does not know in advance how many times loop will be executed .
For example, say we want to know how many times a given number can be divided by 2 before it is less than or equal to 1. example, such as 32, we can say 5 times, but for a given symbolic variable "NUMBER" which represents any number in the world, how many times is not nown before hand. In this case, we could use a while loop to determine that answer.


```python
n = int(input('Enter number of rows : '))

i = 1
while i <= n :
    j = 1
    while j <= i:
        print("*", end = " ")
        j += 1
    print()
    i += 1
```

```python
n = int(input('Enter number of rows : '))
 
i = 1
while i <= n :
    j = n
    while j >= i:
        print("*", end = " ")
        j -= 1
    print()
    i += 1
```

```python
n = int(input('Enter number of rows : '))
 
k = 1
i = 1
while i <= n :
    j = 1
    while j <= i:
        print("{:3d}".format(k), end = " ")
        j += 1
        k += 1
    print()
    i += 1
```
